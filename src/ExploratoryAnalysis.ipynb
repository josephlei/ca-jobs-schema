```python
import pandas as pd
```

```python
df = pd.read_csv('../classifications/schema_alphabetic_final_20161004.csv')
```

```python
# Cleaning Step
df['low'] = df['low'].apply(lambda x: x.strip('$').replace(',',""))
df['high'] = df['high'].apply(lambda x: x.strip('$').replace(",",""))
```

```python
df.low = pd.to_numeric(df.low)
df.high = pd.to_numeric(df.high)
```
